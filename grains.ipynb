```python
df1=df.iloc[:,0:7]
```

```python
df1.sample(5)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,10):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,10),wcss,color='red')
plt.show()
```

```python
kmeans3=KMeans(n_clusters=3)
clusters=kmeans3.fit_predict(df1)
```

```python
clusters
```

```python
KMeans=KMeans(n_clusters=3,n_init=15,random_state=123)
```

### hirerarcy methods

```python
from scipy.cluster.hierarchy import linkage,dendrogram
plt.figure(figsize=[20,10])
merg=linkage(df1,method='ward')
dendrogram(merg,leaf_rotation=90)
plt.title('Dendrogram')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hie_clus=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
cluster2=hie_clus.fit_predict(df1)
df_h=df1.copy(deep=True)
df_h['label']=cluster2
```

```python
from sklearn.cluster import KMeans
kmeans3=KMeans(n_clusters=3)
clusters=kmeans3.fit_predict(df1)
df_k=df1.copy(deep=True)
df_k['label']=clusters
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
ax1=plt.subplot(1,2,1)
plt.title('original classes')
sns.scatterplot(x='A',y='P',hue='target',style='target',data=df,ax=ax1)

ax2=plt.subplot(1,2,2)
plt.title('original classes')
sns.scatterplot(x='A',y='P',hue='label',style='label',data=df_k,ax=ax2)
```

```python
plt.title('original classes')
sns.scatterplot(x='A',y='P',hue='target',style='target',data=df)
plt.show()

plt.title('kmean  classes')
sns.scatterplot(x='A',y='P',hue='label',style='label',data=df_k)
plt.show()


plt.title('hirerchical  classes')
sns.scatterplot(x='A',y='P',hue='label',style='label',data=df_h)
plt.show()
```
